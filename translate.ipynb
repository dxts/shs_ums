```python
from googletrans import Translator
import os
from tqdm import tqdm
```

```python
# txt_folder='nzz_2017_2021_filtered'

# txt_folder='le_temps_2017_2021_filtered'

txt_folder='tagesanzeiger_2017_2021_filtered'
```

```python
translator = Translator()
```

```python

directory = os.fsencode(txt_folder)
    
for file in tqdm(os.listdir(directory)):
    filename = os.fsdecode(file)
     
    with open(f"./{txt_folder}/{filename}", 'r') as inpf:
        origin = inpf.read()

        chunks = [ origin[i : i+4000] for i in range(0, len(origin), 4000) ]

        chunks_translated = []
        for chunk in chunks:
            chunks_translated.append(translator.translate(chunk, dest='en').text)

        dest = ''.join(chunks_translated)

        with open(f"./{txt_folder}_translated/{filename}", 'w') as outf:
            outf.write(dest)
```
